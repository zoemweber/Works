```python
def vel(prof):
    D = 10**prof.log_Dth
    eps_nuc = prof.eps_nuc
    E = prof.energy
    nu_t = E/eps_nuc
    
    v = (sqrt((D)/(nu_t)))
    return v[np.argmax(eps_nuc)]*10**-5
```

```python
def veloc(prof):
    t = prof.star_age*(3.14*10**7)
    r = prof.r_flame
    slope, intercept, r_value, p_value, std_err = stats.linregress(t,r)
    line = (slope*t) + intercept
    return plot(r,t),plot(line,t)
```

```python
def velocity(prof):
    t = prof.star_age[5:]*(3.14*10**7)
    r = prof.r_flame[5:]
    slope, intercept, r_value, p_value, std_err = stats.linregress(t,r)
    line = (slope*t) + intercept
    return slope
```

```python
def v_conduc(x,rho):
    v = 92*((rho/(2*10**9))**.805)*((x/.5)**889)
    return mean(v)
```
