```python
#first import our packages
from numpy import *
%matplotlib inline
from matplotlib.pyplot import *
from glob import *
from astropy.io import ascii
```

```python
#history file containing global quantaties of the star over its simulated lifetime.
history = ascii.read('../../MESA-Web_Job_0205080911/trimmed_history.data',
                    header_start=4,
                    data_start=5)

plot( history['star_age'], history['log_Teff'],'-',color='k' )
xlabel('Time [yr]',fontsize=16)
ylabel(r'$\log(T_{\rm eff}/K) $',fontsize=16)
```

```python
#A look at how the interior composition of the sun changes over time.

files = glob('../MESA-Web_Job_0205080911/profile1[0-9].data')

print files

for i,fn in enumerate(files):
    print i, fn
    
    # read in the data. 
    profile = ascii.read(fn, # the filename we want to read in. 
                    header_start=4,
                    data_start=5)
    
    print "file read... "  
    #Hydrogen mass fraction
    plot(profile['mass'],profile['x'],
        linewidth=2,
        color=cm.jet( float(i)/len(files) ),  
        label=(10+i) )
    #Helium mass fraction
    plot(profile['mass'],profile['y'],'--',
         linewidth=2,
         color=cm.jet( float(i)/len(files) ),
         label=''
         )
    
    
# at the end, we can add some labels 
xlabel(r'Mass $[M_\odot]$',fontsize=20)
ylabel(r'$\log (\rho) \ [g / cm^3]$',fontsize=20)
legend(loc='lower right',labelspacing=0.1)
ylim(0,1)
```

```python
#plotting the  Hertzsprung Russel Diagram

history1 = ascii.read('../MESA-Web_Job_0205080911/trimmed_history.data',
                    header_start=4,
                    data_start=5)

plot( 10**history1['log_Teff'], history1['log_L']  )
#effective temperature
xlabel(r'$T_{\rm eff} [K]$',fontsize=20)
#log Luminosity
ylabel(r'$\log (L/L_\odot)$',fontsize=20)

xlim(6000,3000)
```
